### Utility function for taking input of samples

```python
def input_data(n=None, dtype=float):
    data = []
    i = 0
    while True:
        if i==n: return data
        inp = input(f'Enter element {i+1}: ')
        if not inp and n is None: return data
        data.append(dtype(inp))
        i += 1
```

## 1a: Write a program to implement testing of hypothesis for one sample using Z-test.

```python
from scipy.stats import norm
import numpy as np

def ztest(data, pop_mean, pop_sd, alternative='two-sided', los=0.05):
    n = len(data)
    xbar = np.mean(data)
    Z = (xbar-pop_mean) / (pop_sd/(n**(1/2)))
    
    if alternative=='greater':
        return Z, (pvalue:=norm.sf(Z)), pvalue>los
    elif alternative=='less':
        return Z, (pvalue:=norm.cdf(Z)), pvalue>los
    elif alternative=='two-sided':
        return Z, (pvalue:=2*min(norm.cdf(Z), norm.sf(Z))), pvalue>los
    else:
        raise ValueError("alternative must be 'less', 'greater' or 'two-sided'")
```

```python
data = input_data()
mu = float(input('Enter the population mean: '))
sigma = float(input('Enter the population standard deviation: '))
alt = alt if (alt:=input('Enter the type of test: ')) else 'two-sided'

results = ztest(data, mu, sigma, alternative=alt)
```

```python
print('Sample:', data)
print(f'mu: {mu}, sigma: {sigma}, alternative: {alt}')
results
```

## 1b: Write a program to implement testing of hypothesis of two samples using Z-test.

```python
from scipy.stats import norm
import numpy as np

def ztest(data1, data2, pop_sd1, pop_sd2, alternative='two-sided', los=0.05):
    n1, n2 = len(data1), len(data2)
    xbar1, xbar2 = np.mean(data1), np.mean(data2)
    Z = (xbar1-xbar2) / ((pop_sd1**2/n1)+(pop_sd2**2/n2))**(1/2)
    
    if alternative=='greater':
        return Z, (pvalue:=norm.sf(Z)), pvalue>los
    elif alternative=='less':
        return Z, (pvalue:=norm.cdf(Z)), pvalue>los
    elif alternative=='two-sided':
        return Z, (pvalue:=2*min(norm.cdf(Z), norm.sf(Z))), pvalue>los
    else:
        raise ValueError("alternative must be 'less', 'greater' or 'two-sided'")
```

```python
data1 = input_data()
data2 = input_data()
sigma1 = float(input('Enter the population 1 standard deviation: '))
sigma2 = float(input('Enter the population 2 standard deviation: '))
alt = alt if (alt:=input('Enter the type of test: ')) else 'two-sided'

results = ztest(data1, data2, sigma1, sigma2, alternative=alt)
```

```python
print('Sample 1:', data1)
print('Sample 2:', data2)
print(f'sigma 1: {sigma1}, sigma 2: {sigma2}, alternative: {alt}')
results
```

## 1c: Write a program to implement testing of hypothesis of one sample using t-test

```python
from scipy.stats import t
import numpy as np

def ttest(data, pop_mean, alternative='two-sided', los=0.05):
    n = len(data)
    xbar = np.mean(data)
    s = np.std(data, ddof=1)
    T = (xbar-pop_mean) / (s/(n**(1/2)))
    
    if alternative=='greater':
        return T, (pvalue:=t.sf(T, n-1)), pvalue>los
    elif alternative=='less':
        return T, (pvalue:=t.cdf(T, n-1)), pvalue>los
    elif alternative=='two-sided':
        return T, (pvalue:=2*min(t.cdf(T, n-1), t.sf(T, n-1))), pvalue>los
    else:
        raise ValueError("alternative must be 'less', 'greater' or 'two-sided'")
```

```python
data = input_data()
mu = float(input('Enter the population mean: '))
alt = alt if (alt:=input('Enter the type of test: ')) else 'two-sided'

results = ttest(data, mu, alternative=alt)
```

```python
print('Sample:', data)
print(f'mu: {mu}, alternative: {alt}')
results
```

## 1d: write a program to implement testing of hypothesis of two samples using t-test. 

```python
from scipy.stats import t
import numpy as np

def ttest(data1, data2, alternative='two-sided', los=0.05):
    n1, n2 = len(data1), len(data2)
    xbar1, xbar2 = np.mean(data1), np.mean(data2)
    var1, var2 = np.var(data1, ddof=1), np.var(data2, ddof=1)
    std_comb = (((n1-1)*var1+(n2-1)*var2)/(n1+n2-2))**(1/2)
    T = (xbar1-xbar2) / (std_comb*(((1/n1)+(1/n2))**(1/2)))
    
    if alternative=='greater':
        return T, (pvalue:=t.sf(T, n1+n2-2)), pvalue>los
    elif alternative=='less':
        return T, (pvalue:=t.cdf(T, n1+n2-2)), pvalue>los
    elif alternative=='two-sided':
        return T, (pvalue:=2*min(t.cdf(T, n1+n2-2), t.sf(T, n1+n2-2))), pvalue>los
    else:
        raise ValueError("alternative must be 'less', 'greater' or 'two-sided'")
```

```python
data1 = input_data()
data2 = input_data()
alt = alt if (alt:=input('Enter the type of test: ')) else 'two-sided'

results = ttest(data1, data2, alternative=alt)
```

```python
print('Sample 1:', data1)
print('Sample 2:', data2)
print('alternative:', alt)
results
```

## 1e: Write a program to implement testing of hypothesis of paired samples using t-test.

```python
from scipy.stats import t
import numpy as np

def ttest_dep(data1, data2, alternative='two-sided', los=0.05):
    d = [elem1-elem2 for elem1, elem2 in zip(data1, data2, strict=True)]
    n = len(d)
    dbar = np.mean(d)
    s = np.std(d, ddof=1)
    T = dbar / (s/(n**(1/2)))
    
    if alternative=='greater':
        return T, (pvalue:=t.sf(T, n-1)), pvalue>los
    elif alternative=='less':
        return T, (pvalue:=t.cdf(T, n-1)), pvalue>los
    elif alternative=='two-sided':
        return T, (pvalue:=2*min(t.cdf(T, n-1), t.sf(T, n-1))), pvalue>los
    else:
        raise ValueError("alternative must be 'less', 'greater' or 'two-sided'")
```

```python
data1 = input_data()
data2 = input_data(len(data1))
alt = alt if (alt:=input('Enter the type of test: ')) else 'two-sided'

results = ttest_dep(data1, data2, alternative=alt)
```

```python
print('Sample 1:', data1)
print('Sample 2:', data2)
print('alternative:', alt)
results
```
