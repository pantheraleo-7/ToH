### Utility functions for taking input of samples

```python
def input_data(n=None, dtype=float):
    data = []
    i = 0
    while True:
        if i==n: return data
        inp = input(f'Enter element {i+1}: ')
        if not inp and n is None: return data
        data.append(dtype(inp))
        i += 1

def input_table(r=None, c=None, dtype=float, ragged=False):
    print('Row 1:')
    row1 = input_data(c, dtype=dtype)
    table = [row1]
    c_new = c if ragged else len(row1)
    while True:
        try:
            if len(table)==r: return table
            print(f'Row {len(table)+1}:')
            row = input_data(c_new, dtype=dtype)
            if not row and r is None: return table
            table.append(row)
        except ValueError as e:
            if r is c is None: return table
            else: raise e
```

## 7a: Write a program to implement simple regression analysis.

```python
import numpy as np

class SimpleRegression:
    def __init__(self, degree=1):
        self.deg = degree
    
    def fit(self, x, y):
        x, y = np.asarray(x), np.asarray(y)
        A, b = [], []
        for i in range(self.deg+1):
            A.append([np.sum(x**(i+j)) for j in range(self.deg+1)])
            b.append(np.sum((x**i)*y))
        
        self.coeffs = np.linalg.solve(A, b)
    
    def predict(self, x):
        return np.dot(np.transpose([x])**range(self.deg+1), self.coeffs)
```

```python
linear_regressor = SimpleRegression(degree=1)
quad_regressor = SimpleRegression(degree=2)
```

```python
x = input_data()
y = input_data(len(x))
to_pred = input_data()

linear_regressor.fit(x, y)
quad_regressor.fit(x, y)
```

```python
print('x:', x)
print('y:', y)
print('Values to predict:', to_pred)
```

```python
linear_regressor.predict(to_pred)
```

## 7b: Write a program to implement multiple regression analysis.

```python
import numpy as np

class MultipleRegression:
    def fit(self, X, y):
        X, y = np.c_[np.ones(len(X)), X].T, np.asarray(y)
        A, b = [], []
        for i in range(len(X)):
            A.append([np.sum(X[i]*X[j]) for j in range(len(X))])
            b.append(np.sum(X[i]*y))
        
        self.coeffs = np.linalg.solve(A, b)
    
    def predict(self, X):
        return self.coeffs[0] + np.dot(X, self.coeffs[1:])
```
