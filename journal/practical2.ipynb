### Utility function for taking input of samples

```python
def input_data(n=None, dtype=float):
    data = []
    i = 0
    while True:
        if i==n: return data
        inp = input(f'Enter element {i+1}: ')
        if not inp and n is None: return data
        data.append(dtype(inp))
        i += 1
```

## 2a: Write a program to implement testing of hypothesis for one sample using ꭓ²-test.

```python
from scipy.stats import chi2
import numpy as np

def chi2test(data, pop_var, alternative='two-sided', los=0.05):
    n = len(data)
    s = np.var(data, ddof=1)
    CHI2 = (n-1)*s/pop_var
    
    if alternative=='two-sided':
        return CHI2, (pvalue:=2*min(chi2.cdf(CHI2, n-1), chi2.sf(CHI2, n-1))), pvalue>los
    elif alternative=='greater':
        return CHI2, (pvalue:=chi2.sf(CHI2, n-1)), pvalue>los
    elif alternative=='less':
        return CHI2, (pvalue:=chi2.cdf(CHI2, n-1)), pvalue>los
    else:
        raise ValueError("alternative must be 'less', 'greater' or 'two-sided'")
```

```python
data = input_data()
sigma2 = float(input('Enter the population variance: '))
alt = alt if (alt:=input('Enter the type of test: ')) else 'two-sided'

results = chi2test(data, sigma2, alternative=alt)
```

## 2b: Write a program to implement testing of hypothesis of two samples using F-test.

```python
from scipy.stats import f
import numpy as np

def ftest(data1, data2, alternative='two-sided', los=0.05):
    n1, n2 = len(data1), len(data2)
    s1, s2 = np.var(data1, ddof=1), np.var(data2, ddof=1)
    F = s1/s2
    
    if alternative=='two-sided':
        return F, (pvalue:=2*min(f.cdf(F, n1-1, n2-1), f.sf(F, n1-1, n2-1))), pvalue>los
    elif alternative=='greater':
        return F, (pvalue:=f.sf(F, n1-1, n2-1)), pvalue>los
    elif alternative=='less':
        return F, (pvalue:=f.cdf(F, n1-1, n2-1)), pvalue>los
    else:
        raise ValueError("alternative must be 'less', 'greater' or 'two-sided'")
```

```python
data1 = input_data()
data2 = input_data()
alt = alt if (alt:=input('Enter the type of test: ')) else 'two-sided'

results = ftest(data1, data2, alternative=alt)
```
