## 8a: Write a program to implement time series analysis using moving averages.

```python
def moving_avgs(Y, window):
    msums = [sum(Y[i:i+window]) for i in range(len(Y)-1-window//2)]
    if window%2==0:
        msums = [msums[i]+msums[i+1] for i in range(len(msums)-1)]
        window *= 2
    
    return [msum/window for msum in msums]
```

```python
import numpy as np
T = np.arange(2000, 2100)
Y = np.random.randint(0, 100, size=100)
window = int(input('Enter the window size: '))

ma = moving_avgs(Y, window)
```

```python
import matplotlib.pyplot as plt
plt.plot(T, Y, label='Actual Data')
plt.plot(T[window//2 : -(len(T)-len(ma)-window//2)], 
         ma, label=f'{window}-year MA')
plt.title('Actual Data vs Moving Averages')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.show()
```
